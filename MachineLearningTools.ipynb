```python
import numpy as np
from scipy import stats
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
mean = np.mean(speed)
median = np.median(speed)
mode = stats.mode(speed)
print(mean)
print(median)
print(mode)
print(np.std(speed))
```

```python
perct = np.percentile(speed,20)
print(perct)
```

```python
#linear regression
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats 
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

#simple data plot
plt.scatter(x,y);
plt.xlabel('x-axis---->')
plt.ylabel('y-axis---->')
plt.show()

#use in built linregress function to compute slop and intercepts
slope,intercept,r,p,stderror = stats.linregress(x,y) #in built lingress function
print('slope : ',slope)
print('intercept : ',intercept)
print('r : ',r)
print('p : ',p)
print('stderror : ',stderror)

def calhypothesis(x) :
    return slope*x+intercept

model = list(map(calhypothesis,x))

plt.scatter(x,y)
plt.plot(x,model)
plt.show()
```
