```python
d = dict(chicago['LocationDescription'].value_counts())
d
```

```python
sorted_d = sorted(d.items(),key=operator.itemgetter(1),reverse=True)
count = 0
topFive = []

while count < len(sorted_d):
    if sorted_d[count][0] != 'OTHER':
       topFive.append(sorted_d[count])
    count += 1
    if count == 6:
        break
topFive
```
