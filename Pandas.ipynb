# Slicing

```python
import pandas as pd
df = pd.read_csv('datas.csv')
# use df.column name
print(df.avg_low)
print("\n")
# use df['column name']
print(df['avg_high'])

# number of rows df[start:end]
print(df[0:5])

# df.loc --> location
print(df.loc[0:2,['avg_low','avg_high']])

# using df.iloc it chooses the specific columns 
# Note : while using iloc for colums we use [col1,col2,..]
print(df.iloc[:6,[0,4]])
print(df.iloc[:6,[0,1,4]])
```

# Changing the Values in the File

```python
import pandas as pd 
df = pd.read_csv('datas.csv')

# add a new column 
df['avg_month'] = (df.avg_low + df.avg_high) / 2
print(df)


#add a new row (it doesn't changes the df but returns a new object and assigns to new dataframe object)
newdf = df.append(
    pd.Series(
        ['Neeraj',65,39,100,25,3.06,52.0],
        
        index = df.columns
            ),
        ignore_index=True
    )
print(newdf)
# Always set ignore_index = True while passing dictionary or series

#add a new row using dictionary
newdf = newdf.append( {'month':'Lakshmi',
                     'avg_high':69,
                     'avg_low':21,
                     'record_high':101,
                     'record_low':53,
                     'avg_precipitation':2.01,
                     'avg_month':45
                    },
                  ignore_index=True
                )
print(newdf)


# Adding multiple rows in the dataframe

listOfSeries = [
    pd.Series(['Rahul',80,30,75,22,2.02,55],index=newdf.columns),
    pd.Series(['Mompu',80,30,75,22,2.02,55],index=newdf.columns),
    pd.Series(['Pompu',80,30,75,22,2.02,55],index=newdf.columns)
]

newdf = newdf.append(listOfSeries,ignore_index=True)
print(newdf)

newdf.loc[16,['month']] = 'Popu'
print("\n")
print(newdf)
```

# NaN

```python
'''
    NaN --> Not a Number 
    It is a numeric Datatype used to represent those values that is undefined or unrepresentable.
    ex - 0/0 , sqrt(x) where x < 0 ,log(x) where x < 0 ,sin-1(x)(or cos-1(x)) where x < -1 or x > 1 ,etc
'''
import math
# Nan passes to float as string is not case sensitive 
num = float("Nan")

print(num)
num = float("nan")
print(num)
num = float("NaN")
print(num)
num = float("NAN")
print(num)
num = float("nAN")
print(num)

x = math.nan
y=20.00
print(math.isnan(y))
print(y)
```
